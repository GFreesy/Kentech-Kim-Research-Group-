# Part I. Plotting data

# 1 - Below is an example to plot y=x^2.

```python
# import necessary libraries
# 필요한 라이브러리 import 하기
import matplotlib.pyplot as plt
import numpy as np

# Create an array of x values from -10 to 10
# -10부터 10 범위의 x 값에 대한 배열 생성
x = np.linspace(-10, 10, 100)

# Compute the corresponding y values
# 해당하는 y 값 계산하기
y = x ** 2

# Create a figure and a set of subplots (this is often combined in one step)
# 그래프를 만들고 차트 설정하기 (보통 한 단계로 침)
fig, ax = plt.subplots()

# Plot the data
# 점 찍기
ax.plot(x, y)

# Add labels for x and y axes
# 축 이름 쓰기
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a title
# 차트 제목 쓰기
ax.set_title('$y = x^2$')

# Display the plot
# 차트 띄우기
plt.show()
```

# 2 - Now you choose your favorite function to plot!

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2)

e = np.e
y = e ** (2 * x + 1)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y = e^{2x+1}$')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 11)
y = x ** 2 + 4 * x - 5

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y = x^2+4x-5$')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

pi = np.pi
x = np.linspace(0, 10, 500)
y = 7 * np.cos(pi * x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y = 7cos(πx)$')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5)
y = -1 * np.log10(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$y = −log10(x))$')

fig, ax2 = plt.subplots()
ax2.plot(np.log10(x), y)
ax2.set_xlabel('log x')
ax2.set_ylabel('y')
ax2.set_title('$y = −log10(x))$')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5)
t = [1, 10, 100]
y1 = np.exp(-(x / t[0]))
y2 = np.exp(-(x / t[1]))
y3 = np.exp(-(x / t[2]))

fig, ax = plt.subplots()
ax.plot(x, np.log(y1))
ax.plot(x, np.log(y2))
ax.plot(x, np.log(y3))
ax.set_xlabel('x')
ax.set_ylabel('ln y')
ax.set_title('$y = exp(-x/t)$')
plt.show()
```
